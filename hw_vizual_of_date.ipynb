```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
names_by_year = {}
for year in range (1900, 2010, 10):
    names_by_year[year] = pd.read_csv(
        '/Users/Korolevich/names/yob{}.txt'.format(year),
        names = ['Name', 'Gender', 'Count']
    )
names_all = pd.concat(names_by_year, names = ['Year', 'Pos'])
```

```python
names_all
```

```python
name_dinamics = names_all.groupby([names_all.index.get_level_values(0), 'Name']).sum()
```

```python
name_dinamics.query('Name == ("Ruth", "Robert")').unstack('Name').plot()
```

```python
name_dynamics_cols = name_dinamics.query('Name == ("Ruth", "Robert")').unstack('Name')
```

```python
name_dynamics_cols.plot.bar()
```
